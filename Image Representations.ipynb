# 1. Read 10 images in python, resize each of them a 100 by 100 pixels format;

```python
# Read all 10 images.
import numpy as np
import matplotlib.pyplot as plt
from PIL import Image
from pylab import *
import os

img_folder = r"C:\Users\choin\OneDrive\Desktop\MSIS Winter\522 IT Marketing\image"
img_contents = os.listdir(img_folder)

for image in img_contents: # for loop
    print(image)
    full_path=os.path.join(img_folder, image)
    print(full_path)
    im=Image.open(full_path)
    
    im_rs = im.resize((100,100)) #resize each photos 100 X 100 pixels. 
    display(im_rs)
```

# 2. Convert them to greyscale arrays (no color information needs to be kept, so the arrays will be 2-D);


```python
greyscale_list = []
for image in img_contents: #for loop
    full_path=os.path.join(img_folder, image)
    im=Image.open(full_path)
    print(image)
        
    im_m = array(im) #2-D arrays
    im_grey_img = im.convert('LA')# convert to grey 
    im_grey_m = array(im_grey_img) # convert to grey scale arrays
    im_rs = im_grey_img.resize((100,100))
    greyscale_list.append(im_grey_m) 

    display(im_m)
    display(im_rs)
```

# 3. Flatten the 2-D array to a 1-D array (vectors), draw a histogram to present the intensity value distribution of all of the images

```python
for image in img_contents: #for loop
    full_path=os.path.join(img_folder, image)
    im=Image.open(full_path)

flattened_list = []
for image in greyscale_list: 
    im_v = image.flatten() #Flatten the 2-D array to a 1-D array(vectors).
    print(im_v)
    flattened_list.append(im_v)
    
    
#histogram to present intensity value distribution of images. 
for image in flattened_list:
    figure()
    hist(image, 265)
    plt.axis([0,300,0,20000])
```
